# Team averages


#### Finding a way to calculate team average scores

```python
grouped_teams = s2018.groupby('team')
```

```python
grouped_teams.sum()['PTS']/(len(s2018.groupby(['game_id','team']).mean())/30)
```

#### Averaging everything

```python
game_avgs=pd.DataFrame(grouped_teams.sum())
game_avgs = game_avgs[['AST','BLK','DREB','OREB','PF','PTS','REB','STL','TO',
            'FT_made','FT_attempts','3PT_made','3PT_attempts','FG_made','FG_attempts']]/(len(s2018.groupby(['game_id','team']).mean())/30)
game_avgs['AVG_score']=grouped_teams.sum()['PTS']/(len(s2018.groupby(['game_id','team']).mean())/30)
```
